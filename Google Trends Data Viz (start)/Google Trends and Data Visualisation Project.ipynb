### Bitcoin (BTC) Price v.s. Search Volume

**Challenge**: Create the same chart for the Bitcoin Prices vs. Search volumes. <br>
1. Modify the chart title to read 'Bitcoin News Search vs Resampled Price' <br>
2. Change the y-axis label to 'BTC Price' <br>
3. Change the y- and x-axis limits to improve the appearance <br>
4. Investigate the [linestyles](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.plot.html ) to make the BTC price a dashed line <br>

6. Were big increases in searches for Bitcoin accompanied by big increases in the price?

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Leer y preparar datos de Google Trends
df_btc_search = pd.read_csv('Bitcoin Search Trend.csv')
df_btc_search['MONTH'] = pd.to_datetime(df_btc_search['MONTH'])
df_btc_search.set_index('MONTH', inplace=True)

# Si la columna se llama distinto, ajusta aquÃ­:
# Ejemplo: df_btc_search.rename(columns={'bitcoin': 'BTC_NEWS_SEARCH'}, inplace=True)

# Leer datos del precio diario de Bitcoin
df_btc_price = pd.read_csv('Daily Bitcoin Price.csv')
df_btc_price['DATE'] = pd.to_datetime(df_btc_price['DATE'])

# Resamplear por mes (Ãºltimo valor de cada mes)
df_mensualizado = df_btc_price.resample("M", on='DATE').last()

# --- ConfiguraciÃ³n del grÃ¡fico ---
years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')

plt.figure(figsize=(14,8), dpi=120)
plt.title('Bitcoin News Search vs Resampled Price', fontsize=18)
plt.xticks(fontsize=14, rotation=45)

ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.set_ylabel('BTC Price (USD)', color='#F08F2E', fontsize=14)
ax2.set_ylabel('Search Trend', color='skyblue', fontsize=14)

ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

ax1.set_ylim(bottom=0, top=15000)
ax1.set_xlim([df_mensualizado.index.min(), df_mensualizado.index.max()])

# --- GrÃ¡ficas ---
ax1.plot(df_mensualizado.index, df_mensualizado['CLOSE'],
         color='#F08F2E', linewidth=3, linestyle='--', label='BTC Price')

ax2.plot(df_btc_search.index, df_btc_search['BTC_NEWS_SEARCH'],
         color='skyblue', linewidth=3, marker='o', label='Search Trend')

plt.show()
```

### Unemployement Benefits Search vs. Actual Unemployment in the U.S.

**Challenge** Plot the search for "unemployment benefits" against the unemployment rate.

```python
f_unemployment = pd.read_csv('UE Benefits Search vs UE Rate 2004-19.csv')
df_unemployment['MONTH'] = pd.to_datetime(df_unemployment['MONTH'])

# --- Crear el grÃ¡fico ---
plt.figure(figsize=(14,8), dpi=120)

# Eje principal
ax1 = plt.gca()
ax2 = ax1.twinx()

# TÃ­tulo y etiquetas
plt.title('Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate', fontsize=18)
ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('Unemployment Benefits Web Search', color='skyblue', fontsize=14)
ax2.set_ylabel('FRED U/E Rate', color='orange', fontsize=14)

# Formato del eje X
years = mdates.YearLocator()
years_fmt = mdates.DateFormatter('%Y')
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
plt.xticks(rotation=45, fontsize=12)

# LÃ­mites de los ejes (ajusta segÃºn los datos)
ax1.set_ylim(0, 100)        # escala tÃ­pica para datos de Google Trends
ax2.set_ylim(0, 10)         # tasa de desempleo (%)

# CuadrÃ­cula gris con lÃ­neas discontinuas
ax1.grid(True, which='major', linestyle='--', color='gray', alpha=0.5)

# Graficar
ax1.plot(df_unemployment['MONTH'], df_unemployment['UE_BENEFITS_WEB_SEARCH'], 
         color='skyblue', linewidth=3, label='Search Interest ("Unemployment Benefits")')

ax2.plot(df_unemployment['MONTH'], df_unemployment['UNRATE'], 
         color='orange', linewidth=3, linestyle='--', label='U/E Rate')

# Mostrar leyenda combinada
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

**Challenge**: Calculate the 3-month or 6-month rolling average for the web searches. Plot the 6-month rolling average search data against the actual unemployment. What do you see in the chart? Which line moves first?


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# --- Load and prepare data ---
df_unemployment = pd.read_csv('UE Benefits Search vs UE Rate 2004-19.csv')
df_unemployment['MONTH'] = pd.to_datetime(df_unemployment['MONTH'])

# --- Create 6-month rolling average ---
roll_df = df_unemployment[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=6).mean()

# --- Plot setup ---
plt.figure(figsize=(14,8), dpi=120)
plt.title('Rolling Monthly US "Unemployment Benefits" Web Searches vs UNRATE', fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14, rotation=45)

# Axes and formatting
ax1 = plt.gca()
ax2 = ax1.twinx()

years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')

ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

ax1.set_ylabel('FRED U/E Rate', color='purple', fontsize=16)
ax2.set_ylabel('Search Trend', color='skyblue', fontsize=16)

ax1.set_ylim(bottom=3, top=10.5)
ax1.set_xlim([df_unemployment.MONTH.min(), df_unemployment.MONTH.max()])

# --- Plot lines ---
ax1.plot(df_unemployment.MONTH, roll_df['UNRATE'],
         color='purple', linewidth=3, linestyle='-.', label='U/E Rate (6-mo avg)')
ax2.plot(df_unemployment.MONTH, roll_df['UE_BENEFITS_WEB_SEARCH'],
         color='skyblue', linewidth=3, label='Search Interest (6-mo avg)')

# Grid and legend
ax1.grid(True, linestyle='--', color='gray', alpha=0.5)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
